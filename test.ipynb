```python
import pygame
from pynput.keyboard import Controller, Key
import sys

# --- 初始化 Pygame ---
pygame.init()

# 创建一个小的、不可见的窗口来确保事件能被捕获
# 在 macOS 上，没有窗口的程序可能无法稳定接收输入事件
pygame.display.set_mode((100, 100))
pygame.display.set_caption("Gamepad Mapper")

# --- 初始化手柄 ---
# 检查是否有手柄连接
if pygame.joystick.get_count() == 0:
    print("错误：未检测到任何手柄。")
    pygame.quit()
    sys.exit()

# 假设只用第一个手柄
joystick = pygame.joystick.Joystick(0)
joystick.init()
print("成功连接手柄:", joystick.get_name())

# --- 初始化键盘控制器 ---
keyboard = Controller()

# --- 游戏主循环 ---
clock = pygame.time.Clock()
running = True

print("\n手柄映射已启动...")
print("A 键 (通常是 button 0) -> PageDown")
print("B 键 (通常是 button 1) -> PageUp")
print("按 Ctrl+C 退出程序。")

while running:
    # 处理事件队列
    for event in pygame.event.get():
        # 允许通过点击窗口关闭按钮来退出
        if event.type == pygame.QUIT:
            running = False

        # 监听手柄按键按下事件
        if event.type == pygame.JOYBUTTONDOWN:
            print(f"检测到按键按下: Button {event.button}") # 调试信息：打印按键编号

            if event.button == 0:  # A 键 (在大多数手柄上是0)
                print("A键按下 -> 模拟 PageDown")
                keyboard.press(Key.page_down)
                keyboard.release(Key.page_down)

            elif event.button == 1:  # B 键 (在大多数手柄上是1)
                print("B键按下 -> 模拟 PageUp")
                keyboard.press(Key.page_up)
                keyboard.release(Key.page_up)

    # 控制循环速率，例如每秒30次
    clock.tick(30)

# --- 清理和退出 ---
print("\n程序退出。")
pygame.quit()
sys.exit()
```

```python
import pygame
import time

def main():
    pygame.init()
    pygame.joystick.init()

    joystick_count = pygame.joystick.get_count()
    if joystick_count == 0:
        print("没有检测到手柄，确保已连接")
        return

    # 初始化所有找到的手柄
    joysticks = []
    for i in range(joystick_count):
        joy = pygame.joystick.Joystick(i)
        joy.init()
        joysticks.append(joy)
        print(f"检测到手柄 {i}:")
        print("  名字:", joy.get_name())
        print("  GUID:", joy.get_guid()) # GUID 是更唯一的标识符
        print("  轴 (axes):", joy.get_numaxes())
        print("  按钮 (buttons):", joy.get_numbuttons())
        print("  帽子 (hats):", joy.get_numhats())
        
    print("\n--- 开始监听事件 ---")

    # 设置一个小窗口（对于事件处理是必要的）
    screen = pygame.display.set_mode((200, 100))
    pygame.display.set_caption("Joystick Test")

    try:
        while True:
            # 1. 处理事件队列
            for event in pygame.event.get():
                # print("Got event:", event) # debug 输出所有事件
                if event.type == pygame.QUIT:
                    print("退出事件")
                    return
                
                # --- 手柄事件处理 ---
                # event.instance_id 会告诉你这是哪个手柄实例触发的事件
                
                if event.type == pygame.JOYAXISMOTION:
                    # 添加一个阈值，忽略摇杆的轻微抖动（死区）
                    if abs(event.value) > 0.1:
                        print(f"手柄 {event.instance_id} | 轴 {event.axis} 移动, 值 = {event.value:.3f}")
                elif event.type == pygame.JOYBUTTONDOWN:
                    print(f"手柄 {event.instance_id} | 按钮 {event.button} 按下")
                elif event.type == pygame.JOYBUTTONUP:
                    print(f"手柄 {event.instance_id} | 按钮 {event.button} 松开")
                elif event.type == pygame.JOYHATMOTION:
                    print(f"手柄 {event.instance_id} | 帽子 {event.hat} 移动, 值 = {event.value}")

            # 2. (可选) 轮询所有手柄的状态
            # for i, joy in enumerate(joysticks):
            #     for axis_idx in range(joy.get_numaxes()):
            #         v = joy.get_axis(axis_idx)
            #         if abs(v) > 0.1:
            #             print(f"手柄 {i} (polling) | 轴 {axis_idx} 的值 = {v:.3f}")

            time.sleep(0.02) # 稍微减少延迟以便更快响应
    except KeyboardInterrupt:
        print("用户中断退出")
    finally:
        pygame.quit()

if __name__ == "__main__":
    main()
```

```python
import hid
import struct

class XboxController:
    def __init__(self, vendor_id=0x045E, product_id=0x0B12):
        """
        初始化 Xbox 控制器
        :param vendor_id: 厂商 ID (Microsoft = 0x045E)
        :param product_id: 产品 ID (Xbox Series X Controller USB = 0x0B12)
        """
        self.vendor_id = vendor_id
        self.product_id = product_id
        self.device = hid.device()
        self.device.open(vendor_id, product_id)
        self.device.set_nonblocking(True)

        print("Connected:", self.device.get_manufacturer_string(),
              self.device.get_product_string())

    def read(self):
        """
        从手柄读取一次输入并解析
        :return: dict 格式的状态数据
        """
        data = self.device.read(64)
        if not data:
            return None
        if len(data) < 19:
            return None  # 防御：有时会读到短包

        raw = bytes(data[:19])  # 取前19字节

        report_id = raw[0]
        seq = raw[1] | (raw[2] << 8)
        buttons = raw[3] | (raw[4] << 8)

        lt = raw[5]   # 左扳机
        rt = raw[6]   # 右扳机

        # 小端16位整数 -> 摇杆坐标
        lx, ly, rx, ry = struct.unpack_from("<hhhh", raw, 9)

        return {
            "report_id": report_id,
            "seq": seq,
            "buttons_raw": buttons,
            "buttons": self._decode_buttons(buttons),
            "lt": lt,
            "rt": rt,
            "lx": lx,
            "ly": ly,
            "rx": rx,
            "ry": ry,
        }

    def _decode_buttons(self, bitmask):
        """
        把 bitmask 解析成按钮状态
        :param bitmask: int
        :return: dict {按钮名: True/False}
        """
        mapping = {
            0: "A",
            1: "B",
            2: "X",
            3: "Y",
            4: "LB",
            5: "RB",
            6: "View",
            7: "Menu",
            8: "LS",
            9: "RS",
        }
        return {name: bool(bitmask & (1 << bit)) for bit, name in mapping.items()}


if __name__ == "__main__":
    xbox = XboxController()
    while True:
        state = xbox.read()
        if state:
            print(state)
```

```python
import hid
import struct
import time

class XboxController:
    def __init__(self, vendor_id=0x045E, product_id=0x0B12,
                 use_unsigned_axes=False, report_length=19, nonblocking=True):
        """
        :param use_unsigned_axes: 如果设备把轴数据以 uint16 (0..65535, center ~=32768) 发送，
                                 设置为 True；否则使用 int16（默认 False）。
        """
        self.vendor_id = vendor_id
        self.product_id = product_id
        self.use_unsigned_axes = use_unsigned_axes
        self.report_length = report_length

        self.device = hid.device()
        self.device.open(vendor_id, product_id)
        if nonblocking:
            self.device.set_nonblocking(True)

        print("Connected:", self.device.get_manufacturer_string(),
              self.device.get_product_string())

    def close(self):
        try:
            self.device.close()
        except Exception:
            pass

    def read(self):
        """
        读取并解析一次报告，返回 dict 或 None（无数据）。
        """
        data = self.device.read(64)
        if not data:
            return None
        if len(data) < self.report_length:
            # 有时会读到短包，忽略
            return None

        raw = bytes(data[:self.report_length])

        # 基本字段
        report_id = raw[0]
        seq = raw[1] + (raw[2] << 8)
        buttons = raw[3] + (raw[4] << 8)
        lt = raw[5]
        rt = raw[6]

        # ----- 关键：摇杆从 byte index 10 开始（而不是 9） -----
        if self.use_unsigned_axes:
            ax0, ax1, ax2, ax3 = struct.unpack_from("<HHHH", raw, 10)  # unsigned 16-bit
            lx = self._u16_to_signed(ax0)
            ly = self._u16_to_signed(ax1)
            rx = self._u16_to_signed(ax2)
            ry = self._u16_to_signed(ax3)
        else:
            lx, ly, rx, ry = struct.unpack_from("<hhhh", raw, 10)  # signed 16-bit

        return {
            "raw_bytes": list(raw),
            "report_id": report_id,
            "seq": seq,
            "buttons_raw": buttons,
            "buttons": self._decode_buttons(buttons),
            "lt": lt,
            "rt": rt,
            "lx": lx,
            "ly": ly,
            "rx": rx,
            "ry": ry,
            "lx_norm": self._normalize_axis(lx),
            "ly_norm": self._normalize_axis(ly),
            "rx_norm": self._normalize_axis(rx),
            "ry_norm": self._normalize_axis(ry),
        }

    def _decode_buttons(self, bitmask):
        mapping = {
            0: "A", 1: "B", 2: "X", 3: "Y",
            4: "LB", 5: "RB", 6: "View", 7: "Menu",
            8: "LS", 9: "RS"
        }
        return {name: bool(bitmask & (1 << bit)) for bit, name in mapping.items()}

    def _u16_to_signed(self, v):
        """把 uint16 (0..65535, 中心约 32768) 转成 signed short 范围（-32768..32767）"""
        s = int(v) - 32768
        # 现在 s 理想上在 -32768..32767
        if s > 32767:
            s -= 65536
        return s

    def _normalize_axis(self, v):
        """把 signed short (-32768..32767) 归一化到 -1..1"""
        v = int(v)
        if v < 0:
            return v / 32768.0
        else:
            return v / 32767.0

if __name__ == "__main__":
    try:
        xbox = XboxController(use_unsigned_axes=False)  # 若发现轴中央附近是 ~32768，改成 True
        while True:
            state = xbox.read()
            if state:
                # 简单打印，便于观察
                print(state)
            time.sleep(0.005)
    except KeyboardInterrupt:
        print("退出")
    finally:
        try:
            xbox.close()
        except Exception:
            pass
```

```python
import hid

def open_xbox_controller():
    for dev in hid.enumerate():
        if dev['vendor_id'] == 0x045e and dev['product_id'] == 0x0b12:
            d = hid.device()
            d.open_path(dev['path'])  # 用 path
            return d
    raise RuntimeError("没找到 Xbox 手柄")

# 使用
controller = open_xbox_controller()
print("打开成功:", controller.get_product_string())
```
