```python
from inputs import devices

gamepads = devices.gamepads
if not gamepads:
    print("未检测到手柄")
else:
    pad = gamepads[0]
    print("检测到手柄：", pad)

    while True:
        events = pad.read()
        for event in events:
            print(event.ev_type, event.code, event.state)
```

```python
import pygame
import time

pygame.init()
pygame.joystick.init()

joystick_count = pygame.joystick.get_count()
print(f"检测到 {joystick_count} 个手柄")

if joystick_count == 0:
    exit()

joy = pygame.joystick.Joystick(0)
joy.init()

print("手柄名称:", joy.get_name())

while True:
    pygame.event.pump()  # 刷新事件
    buttons = [joy.get_button(i) for i in range(joy.get_numbuttons())]
    hats = [joy.get_hat(i) for i in range(joy.get_numhats())]
    print("按钮：", buttons)
    print("方向盘：", hats)
    time.sleep(1)
```
