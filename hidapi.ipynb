```python
import hid

def list_hid_devices():
    devices = hid.enumerate()
    print(f"找到 {len(devices)} 个 HID 设备:\n")

    for d in devices:
        vid = d['vendor_id']
        pid = d['product_id']
        product = d['product_string']
        manufacturer = d['manufacturer_string']
        path = d['path'].decode() if isinstance(d['path'], bytes) else d['path']

        print(f"VID:PID = {vid:04x}:{pid:04x}")
        print(f"  设备路径: {path}")
        print(f"  厂商: {manufacturer}")
        print(f"  产品: {product}")
        print(f"  usage_page: {d.get('usage_page')}, usage: {d.get('usage')}")
        print("-" * 40)

if __name__ == "__main__":
    list_hid_devices()
```

```python
import hid

try:
    h = hid.device()
    h.open(0x045e, 0x0b13)  # Microsoft Xbox Wireless Controller
    print("✅ 成功打开设备，说明手柄已连接！")

    print("尝试读取数据...")
    data = h.read(64, timeout_ms=2000)  # 2 秒超时
    print("读取结果:", data)

    h.close()
except Exception as e:
    print("❌ 打开失败，可能手柄未连接:", e)
```
